```python
import json
import requests
import csv
edges = []
capec_csv_file = 'capec.csv'
with open(capec_csv_file, mode='r', encoding='utf-8', errors='replace') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        capec_id = f"CAPEC-{row['ID']}"
        related_weaknesses = row['Related Weaknesses'].split('::')
        related_weaknesses = [cwe for cwe in related_weaknesses if cwe.isdigit()]
        for cwe in related_weaknesses:
            cwe_id = f"CWE-{cwe}"
            edge = {
                "source": cwe_id,
                "target": capec_id,
                "relationship": "cwe_to_capec"
            }
            edges.append(edge)

# Create the final dictionary with "edges" key
output = {
    "edges": edges
}

# Convert the result to JSON string and print
output_json = json.dumps(output, indent=4)
print(output_json)

# Save the result to a JSON file
with open('output.json', 'w') as json_file:
    json.dump(output, json_file, indent=4)
```

```python
import requests
import json
edges=[]
response = requests.get('https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json').json()

# Iterate over all objects in the dataset to create tactic to technique relationships
for item in response['objects']:
    if 'type' in item and item['type'] == 'attack-pattern':
        technique_id = item['external_references'][0]['external_id']
        if 'kill_chain_phases' in item:
            tactics = [phase['phase_name'] for phase in item['kill_chain_phases']]
            for tactic in tactics:
                edge = {
                    "source": tactic,
                    "target": technique_id,
                    "relationship": "tactic_to_technique"
                }
                edges.append(edge)

# Create the final dictionary with "edges" key
output = {
    "edges": edges
}

# Convert the result to JSON string and print
output_json = json.dumps(output, indent=4)
print(output_json)

# Save the result to a JSON file
with open('output.json', 'w') as json_file:
    json.dump(output, json_file, indent=4)
```

tec to tac

```python
import requests
import json

# Initialize an empty list to hold edges
edges = []

# Fetch the ATT&CK dataset from the provided URL
response = requests.get('https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json').json()

# Iterate over all objects in the dataset to create technique to tactic relationships
for item in response['objects']:
    if 'type' in item and item['type'] == 'attack-pattern':
        technique_id = item['external_references'][0]['external_id']
        if 'kill_chain_phases' in item:
            tactics = [phase['phase_name'] for phase in item['kill_chain_phases']]
            for tactic in tactics:
                edge = {
                    "source": technique_id,
                    "target": tactic,
                    "relationship": "technique_to_tactic"
                }
                edges.append(edge)

# Mapping of tactic names to tactic IDs
tactic_to_id = {
    "defense-evasion": "TA0005",
    "privilege-escalation": "TA0004",
    "reconnaissance": "TA0043",
    "resource-development": "TA0042",
    "initial-access": "TA0001",
    "execution": "TA0002",
    "persistence": "TA0003",
    "credential-access": "TA0006",
    "discovery": "TA0007",
    "lateral-movement": "TA0008",
    "collection": "TA0009",
    "command-and-control": "TA0011",
    "exfiltration": "TA0010",
    "impact": "TA0040"
}

# Update edges to replace tactic names with their corresponding IDs
for edge in edges:
    tactic_name = edge["target"]
    if tactic_name in tactic_to_id:
        edge["target"] = tactic_to_id[tactic_name]

# Create the final dictionary with "edges" key
output = {
    "edges": edges
}

# Convert the result to JSON string and print
output_json = json.dumps(output, indent=4)
print(output_json)

# Save the result to a JSON file
with open('tec_to_tac2.json', 'w') as json_file:
    json.dump(output, json_file, indent=4)
```
